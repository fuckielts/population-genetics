Simulate topology of the tree:

```python
def coalescent_tree_taxa(labels):
    nodes = labels.copy()
    # name internal labels: X1, X2, ...
    next_internal = 1
    edges = []
    # remaining node > 1: K >2 lineage remaining:
    while len(nodes) > 1:
        # ramdom choosing two node:
        a, b = random.sample(nodes, 2)
        internal = f"X{next_internal}"
        next_internal += 1
        # let a, b connect to new node:
        edges.append((internal, a))
        edges.append((internal, b))
        # remove because coalescent
        nodes.remove(a)
        nodes.remove(b)
        # nodes can be choose to coalescent next time
        nodes.append(internal)
    root = nodes[0]
    return edges, root


# 2. construct trees: output depth of the tree
def build_tree(edges, root):
    children = {}
    depth = {root: 0}
    for p, c in edges:
        # record connect with parent， (parent)  is key。 (children) is value
        children.setdefault(p, []).append(c)

    def dfs(u):
        # children's depth = parent depth + 1
        for v in children.get(u, []):
            depth[v] = depth[u] + 1
            dfs(v)

    dfs(root)
    return children, depth

def hierarchy_layout(edges, root):
    children, depth = build_tree(edges, root)
    leaves = sorted([n for n in depth if n not in children])
    x = {leaf: i / (len(leaves) - 1) if len(leaves) > 1 else 0.5
         for i, leaf in enumerate(leaves)}
    # from terminal to root node;
    for node in sorted(depth, key=lambda n: depth[n], reverse=True):
        if node in children:
            x[node] = sum(x[c] for c in children[node]) / len(children[node])
    y = {n: -depth[n] for n in depth} # root at the top
    return x, y, children

def draw_tree(edges, root, title="Rooted Tree"):
    x, y, children = hierarchy_layout(edges, root)
    fig, ax = plt.subplots(figsize=(6, 6))
    # Draw edges
    for p, c in edges:
        ax.plot([x[p], x[c]], [y[p], y[c]], linewidth=1, solid_capstyle='round')
    # Draw nodes and labels
    for node in x:
        marker = 's' if node in children else 'o'
        ax.scatter(x[node], y[node], marker=marker, s=300, edgecolors='black')
        ax.text(x[node], y[node], str(node), ha='center', va='center')
    ax.set_title(title, fontsize=15)
    ax.axis('off')
    plt.show()


```

Simulate coalescent tree:

```python
# the non-homogenous possion as waiting time:
def sample_waiting_time(k, t, N0, r):
    C = k * (k - 1) / 2
    lam_t = C * math.exp(r * t) / N0
    x = random.expovariate(1.0)
    return (1 / r) * math.log(1 + (r / lam_t) * x)

# sampling waiting time as the constant rate
def sample_waiting_time_constant(k, N0):
    rate = k * (k - 1) / (2 * N0)
    return np.random.exponential(1 / rate)

# simulate coalescent under contsant N
def coalescent_tree_with_branch_lengths_constant_N(labels, N0):
    nodes         = labels.copy()
    times         = {lab: 0.0 for lab in labels}
    edges         = []
    next_internal = 1
    current_time  = 0.0

    while len(nodes) > 1:
        k = len(nodes)
        w = sample_waiting_time_constant(k, N0) 
        current_time += w

        a, b = random.sample(nodes, 2)
        internal = f"X{next_internal}"; next_internal += 1
        times[internal] = current_time

        edges.append((internal, a, current_time - times[a]))
        edges.append((internal, b, current_time - times[b]))

        nodes.remove(a); nodes.remove(b); nodes.append(internal)

    return edges, nodes[0], times

# simulate coalescent under exponential model
def coalescent_tree_with_branch_lengths_variable_N(labels, N0, r):
    nodes           = labels.copy()
    times           = {lab: 0.0 for lab in labels}   # tip =0
    edges           = []                             # (parent, child, branch_len)
    next_internal   = 1
    current_time    = 0.0

    while len(nodes) > 1:
        k        = len(nodes)
        w = sample_waiting_time(k, current_time, N0, r)
        current_time += w                               

        a, b      = random.sample(nodes, 2)             
        internal  = f"X{next_internal}"; next_internal += 1
        times[internal] = current_time

        
        edges.append((internal, a, current_time - times[a]))
        edges.append((internal, b, current_time - times[b]))

        nodes.remove(a); nodes.remove(b); nodes.append(internal)

    return edges, nodes[0], times                      


# generate Bio.Phylo.Newick.Tree according to edges and time
def edges_to_tree(edges, root, times):
    children = {}
    for p, c, _ in edges:
        children.setdefault(p, []).append(c)

    def build_clade(node, parent_height=None):
        cl = Clade(name=node)
        cl.height = times[node]                        
        if parent_height is None:
            cl.branch_length = 0.0
        else:
            cl.branch_length = parent_height - times[node]

        for ch in children.get(node, []):
            cl.clades.append(build_clade(ch, times[node]))
        return cl

    return Tree(root=build_clade(root))

# transfer function:
def set_branch_from_height(node):
    for child in node.clades:
        child.branch_length = max(node.height - child.height, 0.0)
        set_branch_from_height(child)

def set_height_from_branch(node, acc_height=None):
    if acc_height is None:
        node.height = 0.0                               
    else:
        node.height = acc_height
    for child in node.clades:
        set_height_from_branch(child, node.height + (child.branch_length or 0.0))
```

Simulate mutation: 1. infinite- site model:

```python
# simulate infinite-site mutation model:
def simulate_infinite_sites(labels, edges, times, mu, seq_len):
    children = {}
    for p, c, _ in edges:
        children.setdefault(p, []).append(c)

    branch_lengths = {(p, c): l for p, c, l in edges}
    mutation_map = {label: set() for label in times if label in labels}
    mutation_id = 0
    mutations = []  

    def dfs(node, inherited):
        nonlocal mutation_id
        if node not in children:
            mutation_map[node] = inherited.copy()
            return
        for child in children[node]:
            bl = branch_lengths[(node, child)]
            k = np.random.poisson(mu * seq_len * bl)
            new_mut = set()
            for _ in range(k):
                new_mut.add(mutation_id)
                rel_pos = np.random.uniform(0, 1)
                mutations.append((node, child, rel_pos))
                mutation_id += 1
            dfs(child, inherited | new_mut)

    root = max(times, key=lambda x: times[x])
    dfs(root, set())

    # construct binary matrix
    n = len(labels)
    m = mutation_id
    matrix = np.zeros((n, m), dtype=int)
    label_to_idx = {label: i for i, label in enumerate(labels)}
    for label in labels:
        for site in mutation_map[label]:
            matrix[label_to_idx[label], site] = 1
    return matrix, mutations

```

```python
# assign positions for nodes in a binary tree -- for graph
def layout_binary_tree_topdown(clade, x=0, y=0, x_spacing=1.0):
    
    positions = {}

    def helper(clade, depth):
        if not clade.clades:  # leaf
            positions[clade.name] = (helper.x, clade.height)
            helper.x += x_spacing
        else:
            for child in clade.clades:
                helper(child, depth + 1)
            child_xs = [positions[c.name][0] for c in clade.clades]
            mean_x = sum(child_xs) / len(child_xs)
            positions[clade.name] = (mean_x, clade.height)

    helper.x = x
    helper(clade, 0)
    return positions

# draw a tree with red points as mutation
def draw_symmetric_tree_with_mutations(tree, mutations, title="Coalescent Tree with Mutations"):
    import matplotlib.pyplot as plt

    clade = tree.root
    positions = layout_binary_tree_topdown(clade)

    fig, ax = plt.subplots(figsize=(10, 6))

    def draw_clade(clade):
        x, y = positions[clade.name]
        for child in clade.clades:
            xc, yc = positions[child.name]
            ax.plot([x, xc], [y, yc], color="black")
            draw_clade(child)

    draw_clade(clade)

    # draw nodes:
    for name, (x, y) in positions.items():
        ax.scatter(x, y, facecolor='white', edgecolor='black', s=100, zorder=3)
        ax.text(x, y, name, ha='center', va='center', fontsize=8)

    # mutation red points
    for p, c, rel in mutations:
        if p not in positions or c not in positions:
            continue
        x1, y1 = positions[p]
        x2, y2 = positions[c]
        xm = x1 + rel * (x2 - x1)
        ym = y1 + rel * (y2 - y1)
        ax.scatter(xm, ym, color='red', s=30, zorder=4)
    
    y_values = list(pos[1] for pos in positions.values())
    ymin, ymax = min(y_values), max(y_values)
    ax.set_ylim(ymax + 5, ymin - 5) 
    
    # make root on top (past → present)
    ax.set_ylim(ax.get_ylim()[::-1])
    ax.set_title(title)
    ax.set_ylabel("Time (from past to present)")
    ax.set_xticks([])
    plt.tight_layout()
    plt.show()
```

Simulate mutation using JC69 model

```python
# simulate JC69 model:
def simulate_jc69(labels, edges, times, mu, seq_len):
    BASES = ['A', 'C', 'G', 'T']
    def random_base_exclude(b):
        return random.choice([x for x in BASES if x != b])
    # parent→children mapping + branch_lengths
    children = {}
    for p, c, _ in edges:
        children.setdefault(p, []).append(c)
    branch_lengths = {(p, c): l for p, c, l in edges}
    root = max(times, key=times.get)
    # initialize sequence: (without mutation)
    seqs = {root: [random.choice(BASES) for _ in range(seq_len)]}
    mutations = []  # (parent, child, rel, site, from, to)  rel: relative position in branch:
    def dfs(node):
        for child in children.get(node, []):
            # extract parent, children, and branch length.
            bl = branch_lengths[(node, child)]
            parent_seq = seqs[node]
            child_seq = parent_seq[:]
            # every site is a independent JC69 mutation：
            for i in range(seq_len):
                t = 0.0
                base = parent_seq[i]
                while True:
                    # waiting for the next mutation event
                    wait = random.expovariate(3 * mu)
                    t += wait
                    # record mutation point in one branch length:
                    if t > bl:
                        break
                    # mutate process
                    new_base = random_base_exclude(base)
                    mutations.append((node, child, t / bl, i, base, new_base))
                    base = new_base
                child_seq[i] = base
            seqs[child] = child_seq
            dfs(child)
    dfs(root)
    return seqs, mutations
```

```python
# draw tree under JC69 model 
def draw_symmetric_tree_for_site(tree, site, seqs, jc_muts, title="Symmetric JC69 Tree at site"):
    from matplotlib.patches import Patch
    cmap = {'A': 'red', 'C': 'blue', 'G': 'green', 'T': 'orange'}

    # Binary layoutL
    positions = layout_binary_tree_topdown(tree.root)
    fig, ax = plt.subplots(figsize=(10, 6))

    # draw edges:
    def draw_edges(clade):
        x0, y0 = positions[clade.name]
        for ch in clade.clades:
            x1, y1 = positions[ch.name]
            ax.plot([x0, x1], [y0, y1], lw=2, color='#444444')
            draw_edges(ch)

    draw_edges(tree.root)

    # sequence 
    for name, (x, y) in positions.items():
        base = seqs[name][site]
        ax.scatter(x, y, s=100, color=cmap[base], edgecolor='black', zorder=3)
        ax.text(x, y, f"{name}\n{base}", fontsize=7, ha='center', va='center', color='white')

    # mutation positions
    for p, c, rel, s, fr, to in jc_muts:
        if s != site:
            continue
        if p not in positions or c not in positions:
            continue
        x1, y1 = positions[p]
        x2, y2 = positions[c]
        xm = x1 + rel * (x2 - x1)
        ym = y1 + rel * (y2 - y1)
        ax.scatter(xm, ym, s=50, color=cmap[to], edgecolor='black', alpha=0.9, zorder=6)
        ax.text(xm, ym + 0.015, f"{fr}→{to}", fontsize=6, ha='center', va='bottom')

    legend_elements = [Patch(facecolor=c, edgecolor='black', label=b) for b, c in cmap.items()]
    ax.legend(handles=legend_elements, title="Base", loc="upper right")
    ax.set_ylabel("Time (from past to present)")
    ax.set_xticks([])
    ax.set_title(f"{title}", fontsize=13)
    plt.tight_layout()
    plt.show()

```

```python
labels = ["A", "B", "C", "D", "E","F","G"]
N0 = 1000
r = 0.01
mu = 1e-4
seq_length = 2000

# simulate coalescent tree & JC69 mutation
edges, root, times = coalescent_tree_with_branch_lengths_variable_N(labels, N0, r)
seqs, jc_muts = simulate_jc69(labels, edges, times, mu, seq_length)
tree = edges_to_tree(edges, root, times)

# choose a mutation position
def pick_one_mutated_site(jc_muts):
    sites = list(set(m[3] for m in jc_muts))
    return random.choice(sites) if sites else None

site = pick_one_mutated_site(jc_muts)
#print(f"Picked mutated site: {site}")

draw_symmetric_tree_for_site(
    tree=tree, site=site,
    seqs=seqs, jc_muts=jc_muts,
    title=f"Mutation Tree at Site {site}"
)
```

Using msprime to simulate data：using infinite-site model：

simulate constant model:

```python
def simulate_constant(N0,n,seq_len,mu,random_seed,):
    dem = msprime.Demography()
    dem.add_population(name="pop", initial_size=N0)

    ts = msprime.sim_ancestry(
        samples=[msprime.SampleSet(n, population="pop", ploidy=1)],
        demography=dem,
        sequence_length=seq_len,
        recombination_rate=0,
        random_seed=random_seed
    )
    ts_tree = ts.first()
    mts = msprime.sim_mutations(
        ts,
        rate=mu,
        model=msprime.InfiniteSites(),
        random_seed=random_seed
    )
    mu_tree = mts.first()
    num_nodes = len(mu_tree.preorder())
    node_times = np.array([mu_tree.time(u) for u in range(num_nodes)])
    internal_nodes = [u for u in range(num_nodes) if mu_tree.num_children(u) > 0]
    coalescent_times = np.sort(node_times[internal_nodes])
    return coalescent_times,mts,ts
```

simulate exponential model:

```python
def simulate_exponential(growth_rate,n,seq_len,mu,N0,random_seed,):

    dem = msprime.Demography()
    dem.add_population(name="pop", initial_size=N0)
    dem.add_population_parameters_change(time=1e-8, growth_rate=growth_rate,
                                         population="pop")

    ts = msprime.sim_ancestry(
        samples=[msprime.SampleSet(n, population="pop", ploidy=1)],
        demography=dem,
        sequence_length=seq_len,
        recombination_rate=0,
        random_seed=random_seed
    )
    ts_tree = ts.first()
    #ts_tree.draw_svg(path="tree_exponential.svg")
    tree_height = ts_tree.time(ts_tree.root)
    
    mts = msprime.sim_mutations(
        ts,
        rate=mu,
        model=msprime.InfiniteSites(),
        random_seed=random_seed
    )
    mu_tree = mts.first()
    #mu_tree.draw_svg(path="mu_tree_exponential.svg")
    #print(f"  Mutations: {mts.num_mutations}, variable sites: {mts.num_sites}")

    num_nodes = len(mu_tree.preorder())
    node_times = np.array([mu_tree.time(u) for u in range(num_nodes)])
    internal_nodes = [u for u in range(num_nodes) if mu_tree.num_children(u) > 0]
    coalescent_times = np.sort(node_times[internal_nodes])

    return coalescent_times, mts, ts

```

```python
def plot_exponential_and_constant(N0, r, t_max=1000, num_points=500):
    t = np.linspace(0, t_max, num_points)
    Ne_exp = N0 * np.exp(-r * t)      
    Ne_const = np.full_like(t, N0)    
    
    plt.figure(figsize=(8, 5))
    plt.plot(t, Ne_exp, label=f"Exponential: $N(t) = {N0} \cdot e^{{-{r}t}}$", color="C1")
    plt.plot(t, Ne_const, "--", label=f"Constant: $N(t) = {N0}$", color="C0")
    
    plt.xlabel("Backward time (from present to past)")
    plt.ylabel("Effective Population Size")
    plt.title("Exponential vs Constant Population History")
    plt.ylim(0, N0 * 1.2)
    plt.grid(True, alpha=0.5)
    plt.legend()
    plt.tight_layout()
    plt.show()


plot_exponential_and_constant(N0=500, r=0.01)
```

simulate piecewise function:

```python
def simulate_piecewise(
        #output_fasta,
        N0, N1, N2, t1,t2,
        n=n,
        seq_len=seq_length,
        mu=mu,
        random_seed=60,
        ):
    dem = msprime.Demography()

    dem.add_population(name="pop", initial_size=N0)
    dem.add_population_parameters_change(time=t1, initial_size=N1, population="pop")
    dem.add_population_parameters_change(time=t2, initial_size=N2, population="pop")

    ts = msprime.sim_ancestry(
        samples=[msprime.SampleSet(n, population="pop", ploidy=1)],
        demography=dem,
        sequence_length=seq_len,
        recombination_rate=0,
        random_seed=random_seed
    )
    ts_tree = ts.first()

    tree_height = ts_tree.time(ts_tree.root)

    mts = msprime.sim_mutations(
        ts,
        rate=mu,
        model=msprime.InfiniteSites(),
        random_seed=random_seed
    )
    mu_tree = mts.first()
    #print(f"  Mutations: {mts.num_mutations}, variable sites: {mts.num_sites}")

    num_nodes = len(mu_tree.preorder())
    node_times = np.array([mu_tree.time(u) for u in range(num_nodes)])
    internal_nodes = [u for u in range(num_nodes) if mu_tree.num_children(u) > 0]
    coalescent_times = np.sort(node_times[internal_nodes])

    return coalescent_times,mts,ts

```

```python
def plot_population_piece_vs_constant(N0, N1, N2, t1, t2, time_max=1000):
   
    times = [0, t1, t2, time_max]
    sizes = [N0, N1, N2, N2]
    
    # constant population (baseline)
    const_times = [0, time_max]
    const_sizes = [N0, N0]
    
    plt.figure(figsize=(8, 5))
    
   
    plt.step(times, sizes, where="post", 
             label=f"Piecewise: {N0}→{N1}→{N2}", color="C1")
    
    
    plt.plot(const_times, const_sizes, "--", 
             label=f"Constant: $N(t)={N0}$", color="C0")
    
    plt.xlabel("Time (from present to past)")
    plt.ylabel("Effective Population Size (Ne)")
    plt.title("Piecewise vs Constant Population History")
    plt.ylim(0, max(max(sizes), N0) * 1.2)
    plt.xlim(0, time_max)
    plt.grid(True, alpha=0.5)
    plt.legend()
    plt.tight_layout()
    plt.show()


plot_population_piece_vs_constant(N0=500, N1=50, N2=500, t1=80, t2=160)


```

```python
def simulate_decay(
        N0, N1, t1,n,seq_len,mu,random_seed=60,):
    dem = msprime.Demography()

    dem.add_population(name="pop", initial_size=N0)
    dem.add_population_parameters_change(time=t1, initial_size=N1, population="pop")

    ts = msprime.sim_ancestry(
        samples=[msprime.SampleSet(n, population="pop", ploidy=1)],
        demography=dem,
        sequence_length=seq_len,
        recombination_rate=0,
        random_seed=random_seed
    )
    ts_tree = ts.first()

    mts = msprime.sim_mutations(
        ts,
        rate=mu,
        model=msprime.InfiniteSites(),
        random_seed=random_seed
    )
    mu_tree = mts.first()
    #print(f"  Mutations: {mts.num_mutations}, variable sites: {mts.num_sites}")

    num_nodes = len(mu_tree.preorder())
    node_times = np.array([mu_tree.time(u) for u in range(num_nodes)])
    internal_nodes = [u for u in range(num_nodes) if mu_tree.num_children(u) > 0]
    coalescent_times = np.sort(node_times[internal_nodes])
    #write_fasta(mts, output_fasta, n=n)
    return coalescent_times,mts,ts

def plot_population_change_decay(N0,N1,t1):
    time_max=1000
    times = [0, t1,  time_max]
    sizes = [N0, N1, N1]
    plt.figure(figsize=(8, 5))
    plt.step(times, sizes, where="post", label="Decay Population Size")
    plt.ylabel("Effective Population Size (Ne)")
    plt.title("Decay Population History")
    plt.ylim(0, max(sizes) * 1.2)
    plt.xlim(0, time_max)
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()



```

```python
# calculate Tajima's D from mutation
def tajima_d_from_mts(mts):
    n = mts.num_samples
    if n < 2:
        return float('nan'), 0.0, 0

    S = 0          # number of segregating sites
    pi_sum = 0.0   # sum over sites of pairwise diversity contribution
    for var in mts.variants():
        geno = var.genotypes           
        counts = np.bincount(geno, minlength=len(var.alleles))
        derived_count = int(np.sum(counts[1:])) 
        if derived_count == 0 or derived_count == n:
            continue  
        S += 1
        pi_sum += (2.0 * derived_count * (n - derived_count)) / (n * (n - 1))
    if S == 0:
        return float('nan'), 0.0, 0
    # Harmonic sums
    a1 = sum(1.0 / i for i in range(1, n))
    a2 = sum(1.0 / (i * i) for i in range(1, n))
    # Watterson's theta_W
    theta_w = S / a1
    # Variance constants (Tajima 1989)
    b1 = (n + 1) / (3 * (n - 1))
    b2 = 2 * (n**2 + n + 3) / (9 * n * (n - 1))
    c1 = b1 - (1 / a1)
    c2 = b2 - ((n + 2) / (a1 * n)) + (a2 / (a1**2))
    e1 = c1 / a1
    e2 = c2 / (a1**2 + a2)
    var_d = e1 * S + e2 * S * (S - 1)
    if var_d <= 0:
        return float('nan'), pi_sum, S
    D = (pi_sum - theta_w) / math.sqrt(var_d)
    return D, pi_sum, S
```

```python

# replicate to get distribution of Tajima's D
def tajima_d_replicates_mean(sim_func, seed_list):
    D_reps = []
    pi_reps = []
    theta_reps = []

    for seed in seed_list:
        result = sim_func(random_seed=seed)
        mts = result[1]                         
        D, pi_val, S = tajima_d_from_mts(mts)   
        n = mts.num_samples                     

        a1 = sum(1.0 / i for i in range(1, n))  
        theta_w = S / a1 if a1 > 0 else float('nan')

        if not math.isnan(D):
            D_reps.append(D)
        pi_reps.append(pi_val)
        theta_reps.append(theta_w)

    D_mean     = np.nanmean(D_reps) if D_reps else float('nan')
    D_sd       = np.nanstd(D_reps, ddof=1) if len(D_reps) > 1 else float('nan')
    pi_mean    = float(np.mean(pi_reps)) if pi_reps else 0.0
    pi_sd      = float(np.std(pi_reps, ddof=1)) if len(pi_reps) > 1 else float('nan')
    theta_mean = float(np.mean(theta_reps)) if theta_reps else 0.0
    theta_sd   = float(np.std(theta_reps, ddof=1)) if len(theta_reps) > 1 else float('nan')

    return {
        "D_mean":   D_mean,
        "D_sd":     D_sd,
        "D_reps":   D_reps,
        "pi_mean":  pi_mean,
        "pi_sd":    pi_sd,
        "theta_w_mean": theta_mean,
        "theta_w_sd":   theta_sd,
    }
```

```python

# calsulate table
def tajima_d_replicates(sim_func, seed_list, tajima_func=tajima_d_from_mts):
    recs = []

    for seed in seed_list:
        out = sim_func(random_seed=seed)          
        mts = out[1] if isinstance(out, tuple) else out

        D, pi_val, S = tajima_func(mts)
        n = mts.num_samples                       
        a1 = sum(1.0 / i for i in range(1, n))    
        theta_w = S / a1 if a1 > 0 else float('nan')

        recs.append({"seed": seed,
                     "D": D,
                     "pi": pi_val,
                     "theta_w": theta_w})

    df = pd.DataFrame.from_records(recs)

    # -------- summary --------
    summary = {}
    for col in ["D", "pi", "theta_w"]:
        arr = df[col].to_numpy()
        summary[f"{col}_mean"] = float(np.nanmean(arr))
        summary[f"{col}_sd"]   = float(np.nanstd(arr, ddof=1))
        summary[f"{col}_p2.5"] = float(np.nanpercentile(arr, 2.5))
        summary[f"{col}_p97.5"] = float(np.nanpercentile(arr, 97.5))
    summary["n_reps"] = len(df)

    return df, summary

def tajima_summary_table(summary_dict_by_label, roundto=4):
    rows = []
    for label, s in summary_dict_by_label.items():
        rows.append({
            "model":   label,
            "D_mean":  s.get("D_mean",   np.nan),
            "D_sd":    s.get("D_sd",     np.nan),
            "D_p2.5":  s.get("D_p2.5",   np.nan),
            "D_p97.5": s.get("D_p97.5",  np.nan),
            "theta_w_mean": s.get("theta_w_mean", np.nan),
            "theta_w_sd":   s.get("theta_w_sd",   np.nan),
            "pi_mean": s.get("pi_mean",  np.nan),
            "pi_sd":   s.get("pi_sd",    np.nan),
            "n_reps":  s.get("n_reps",   np.nan),
        })

    df = pd.DataFrame(rows).set_index("model")

    if roundto is not None:
        numeric_cols = df.select_dtypes(include=[np.number]).columns
        df[numeric_cols] = df[numeric_cols].round(roundto)

    return df



def plot_empirical_distribution(df, col="D", bins=50, title=None, show_mean=True, ax=None, color=None):
    data = df[col].dropna().to_numpy()

    if ax is None:
        fig, ax = plt.subplots()

    ax.hist(data, bins=bins, density=True, color=color, edgecolor='black', alpha=0.7)

    if show_mean:
        m = np.nanmean(data)
        ax.axvline(m, color="red", linestyle="--", linewidth=1.5)

    ax.set_xlabel(col)
    ax.set_ylabel("Density")
    if title:
        ax.set_title(title)



def plot_multiple_models_grid(model_df_dict, col="D", bins=50):
    import matplotlib.pyplot as plt

    n_models = len(model_df_dict)
    fig, axes = plt.subplots(2, 2, figsize=(10, 8))
    axes = axes.flatten() 

    colors = plt.cm.Set2.colors  

    for i, (name, df) in enumerate(model_df_dict.items()):
        plot_empirical_distribution(
            df, col=col, bins=bins, title=name,
            show_mean=True, ax=axes[i], color=colors[i % len(colors)]
        )

    for j in range(i+1, len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()
    plt.show()

# plot boxplot
def boxplot_models(model_df_dict, col="D"):
    import pandas as pd
    df_combined = pd.concat([
        df[[col]].assign(model=name) for name, df in model_df_dict.items()
    ])
    sns.boxplot(data=df_combined, x="model", y=col)
    plt.title(f"Boxplot of Tajima's D by model")
    plt.grid(True)
    plt.tight_layout()
    plt.show()


```

```python
# calculate SFS:
def repeated_average_sfs(sim_func, seed_list):
    sfs_list = []
    for seed in seed_list:
        result = sim_func(random_seed=seed)
        mts = result[1]                     # assume mts is the second return
        sfs = mts.allele_frequency_spectrum(
            mode="site", polarised=True, span_normalise=False
        )
        sfs_list.append(sfs)
    sfs_array = np.stack(sfs_list, axis=0)  # shape (R, n+1)
    mean_sfs = np.mean(sfs_array, axis=0)
    return mean_sfs, sfs_array
def plot_sfs_comparison(mean_sfs1, mean_sfs2,
                        labels=('Model 1', 'Model 2'),
                        normalize=False,
                        figsize=(10, 6),
                        alpha=0.6,
                        width=0.4):
   
    # Prepare data (skip bin 0)
    x = np.arange(1, len(mean_sfs1))
    y1 = mean_sfs1[1:].astype(float)
    y2 = mean_sfs2[1:].astype(float)

    if normalize:
        y1 /= y1.sum()
        y2 /= y2.sum()
        ylabel = 'Proportion of sites'
        title = 'Normalized Average SFS Histograms'
    else:
        ylabel = 'Average number of sites'
        title = 'Site Frequency Spectrum'

    # offset bars so they overlap side-by-side
    x1 = x - width / 2
    x2 = x + width / 2

    plt.figure(figsize=figsize)
    plt.bar(x1, y1, width=width, alpha=alpha, label=labels[0])
    plt.bar(x2, y2, width=width, alpha=alpha, label=labels[1])
    plt.xlabel('Derived allele count')
    plt.ylabel(ylabel)
    plt.title(f'{title}: {labels[0]} vs {labels[1]}')
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
# NNI:
from copy import deepcopy
def normalize_root_height(tree, target_height):
    scale = target_height / tree.root.height
    def scale_heights(node):
        node.height *= scale
        for c in node.clades:
            scale_heights(c)
    scale_heights(tree.root)

def height_to_branch(node):
    for child in node.clades:
        child.branch_length = max(node.height - child.height, 0.0)
        height_to_branch(child)

def reassign_heights_local(parent, child, root_height, eps_frac=0.05):
    a, b = child.clades
    low  = max(a.height, b.height) + 1e-6
    high = parent.height - 1e-6
    child.height = random.uniform(low, high)

    other = parent.clades[1] if parent.clades[0] is child else parent.clades[0]
    low_p = max(other.height, child.height) + 1e-6
    eps = eps_frac * root_height
    old_height = parent.height
    upper = min(old_height + eps, root_height)
    lower = max(low_p, old_height - eps)
    parent.height = random.uniform(lower, upper)

def enforce_ultrametric(node):
    if node.is_terminal():                    # leaf
        return node.height
    child_h = [enforce_ultrametric(c) for c in node.clades]
    node.height = max(child_h) + 1e-6         # ensure parent > children
    return node.height

def propose_nni_ultrametric(tree):
    prop = deepcopy(tree)
    original_root_height = tree.root.height   
    # find (parent, idx, child)
    cand = [(p, i, c)
            for p in prop.get_nonterminals(order="preorder")
            for i, c in enumerate(p.clades) if len(c.clades) == 2]
    if not cand:
        return prop

    parent, idx, child = random.choice(cand)
    sib        = parent.clades[1 - idx]
    b1, b2     = child.clades

    # topology change
    if random.random() < 0.5:
        parent.clades[1 - idx] = b1
        child.clades[0]        = sib
    else:
        parent.clades[1 - idx] = b2
        child.clades[1]        = sib

    reassign_heights_local(parent, child, original_root_height)  
    enforce_ultrametric(prop.root)
    normalize_root_height(prop, original_root_height)
    return prop
```

```python
def build_parent_map(tree):
    parent_map = {}
    for parent in tree.find_clades(order="preorder"):
        for child in parent.clades:
            parent_map[child] = parent
    return parent_map

def is_ancestor(u, v, parent_map):
    cur = parent_map.get(v)
    while cur:
        if cur is u:
            return True
        cur = parent_map.get(cur)
    return False



```

```python
from copy import deepcopy
import random
from Bio.Phylo.Newick import Clade

EPS = 1e-6  

def propose_spr_standard_spr(tree, eps_frac=0.05):
    
    prop = deepcopy(tree)
    root = prop.root
    root_height = getattr(root, "height", 0.0)

    parent_map = build_parent_map(prop)

    # choose internal node :
    prune_candidates = [n for n in parent_map if not n.is_terminal()]
    if not prune_candidates:
        return prop
    prune_node = random.choice(prune_candidates)
    parent_of_prune = parent_map[prune_node]
    original_edge = (parent_of_prune, prune_node)

    # cut prune_node
    parent_of_prune.clades.remove(prune_node)

    # choose another branch) 
    parent_map = build_parent_map(prop)
    attach_edges = []
    for p in prop.find_clades(order="preorder"):
        for c in p.clades:
            if (p, c) == original_edge:
                continue
            if is_ancestor(prune_node, p, parent_map):  
                continue
            attach_edges.append((p, c))
    if not attach_edges:
        return prop
    attach_parent, attach_child = random.choice(attach_edges)

    # 4) insert new_internal
    if getattr(prune_node, "name", None) is None:
        prune_node.name = "PRUNE"
    new_internal = Clade(name="new node")
    idx = attach_parent.clades.index(attach_child)
    attach_parent.clades[idx] = new_internal
    new_internal.clades = [attach_child, prune_node]

    # change branch length to ensure ultrametric  tree:
    # new_internal.height in the range of  (max(children), attach_parent.height)
    child_heights = [getattr(ch, "height", 0.0) for ch in new_internal.clades]
    low_new  = max(child_heights) + 10*EPS
    high_new = getattr(attach_parent, "height", root_height) - 10*EPS
    new_internal.height = (
    0.5 * (low_new + high_new) if high_new <= low_new 
    else random.uniform(low_new, high_new))
    # prune_node.height in the range of (max(children of prune_node), new_internal.height)
    if prune_node.clades:
        max_child = max(getattr(ch, "height", 0.0) for ch in prune_node.clades)
    else:
        max_child = 0.0
    low_p  = max_child + 10*EPS
    high_p = new_internal.height - 10*EPS
    eps_move = eps_frac * max(root_height, 1.0)
    lower = max(low_p, getattr(prune_node, "height", low_p) - eps_move)
    upper = min(high_p, getattr(prune_node, "height", low_p) + eps_move)
    if upper <= lower:
        prune_node.height = low_p + 1e-5
    else:
        prune_node.height = random.uniform(lower, upper)
    enforce_ultrametric(prop.root)
    normalize_root_height(prop, root_height)

    return prop

def ensure_internal_names(tree, prefix="X"):
    k = 1
    for cl in tree.find_clades(order="postorder"):
        if not cl.is_terminal() and not getattr(cl, "name", None):
            cl.name = f"{prefix}{k}"
            k += 1

```
